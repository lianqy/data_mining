```python
data = pd.read_csv('./train.csv')
X = data.T
X = X[1:385]
X = X.T
Y = data['reference']
```

```python
tdata = pd.read_csv('./test.csv')
pre_X = tdata.T
pre_X = pre_X[1:385]
pre_X = pre_X.T
```

声明KNN的回归器

```python
knn = neighbors.KNeighborsRegressor(n_neighbors=2, n_jobs=24)
```

```python
knn.fit(X,Y)
```
